```python
###LOGIT
dataForLogit = pd.read_csv('/Users/xiaominli/Documents/habitPCS/habitPCShospital/onlyPreventixData.csv')
```

```python
dataForLogit['whetherBeforePrevention'] = pd.to_datetime(dataForLogit['datetime'])>pd.to_datetime(dataForLogit['preventionDate'])
```

```python
dataPerson = dataForLogit[dataForLogit['hospital_tag']==6]
dataPerson.columns
```

```python
cols=['whetherBeforePrevention'] 
X=dataPerson[cols]
y=dataPerson['compliant']
logit_model=sm.Logit(y,X)
result=logit_model.fit()
print(result.params)
```

```python
allIds = dataForLogit['hospital_tag'].unique()
useIds = list()
coefs = list()
for personId in allIds:
    dataPerson = dataForLogit[dataForLogit['hospital_tag']==personId]
    if dataPerson.shape[0]<30:
        continue
    y, X = dmatrices('compliant ~ whetherBeforePrevention', dataPerson, return_type = 'dataframe')
    model = LogisticRegression(fit_intercept = True, C = 1e9)
    mdl = model.fit(X, y)
    coefs.append(model.coef_[0][1])
    useIds.append(personId)
```

```python
d = {'userId': useIds, 'coefs': coefs}
```

```python
df = pd.DataFrame(data=d)
df
```
